```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

print('Imports complete.')
```

```python
df = pd.read_csv('./primes.csv')
df.head()
```

```python
# We are going to try to predict the 4th prime in the series
X = df.loc[:, df.columns != 'p4']
y = df['p4']


print(X.head())
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Since we predicting for a number, not a classification, we need a regressor model
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
guesses = rf.predict(X_test)


print('p1\tp2\tp3\tactual\tguess\tdiff')
for i in range(len(guesses)):
    print('{0}\t{1}\t{2}\t{4}\t{3}\t{5:.2f}'.format(X_test.iloc[i]['p1'], X_test.iloc[i]['p2'], X_test.iloc[i]['p3'], guesses[i], y_test.iloc[i], guesses[i]-y_test.iloc[i]))
```
